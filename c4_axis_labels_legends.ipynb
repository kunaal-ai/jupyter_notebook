```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
min_temp = [0,6,4,7,5,9,6]
max_temp = [12,20,18,25,22,30,25]
avg_temp = [1,2,3,4,5,6,7]
```

```python
plt.xlabel('Days')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(days, max_temp, 'r--', label='Max')
plt.plot(days, min_temp, 'g-.', label='Min')
plt.plot(days, avg_temp, label='Avg')

plt.legend(loc='best', shadow = 'true', fontsize='large')
plt.grid()
# fontsize = large / small
# Location String	Location Code --> https://matplotlib.org/api/legend_api.html?highlight=legend#module-matplotlib.legend
# After adding labels need to add legend() property
# loc --> 'upper right', 'best' (default)
# shadow = true --> legends shadow
```
